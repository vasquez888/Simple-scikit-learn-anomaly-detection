Before we play further with anomaly detection, let's figure out how to create some data. Incidentally, I believe the seaborn package will probably aide us most visualizing the data. It works nicely with pandas dataframes.<BR>

The make_blobs function is pretty straight forward. It creates clusters of normally distributed data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

<h1>make_blobs</h1>
This function outputs two arrays--the blobs and the labels for the blobs. I'll only be dealing with 2 features as they're easiest to vizualize and think about.<BR>

One thing to note is that you can change the default standard deviation of 1 with the cluster_std paramter.

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=0)
print('X = ', X[0:6, :]); print('y= ', y[0:6])
fig, ax = plt.subplots(1, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=.5)
plt.show()
```

The seaborn package provides better visualizations that are easier to produce, however. We'll use that from this point onward.<BR>
<h1>make_classification</h1>
This function is a little more complicated. It's similar to make_blobs in that it makes normally distributed data but it provides a good degree more of control over the clusters per class and which features are "informative". Frankly, I find the documentation to be non-trivial to understand. You can find some good examples <A HREF="http://scikit-learn.org/stable/auto_examples/datasets/plot_random_dataset.html"> here.</A> Playing with the parameters should give the intuition needed to set them.

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=400, n_features=2, random_state=0,
                           n_informative=2, n_redundant=0,
                           n_classes=3, n_clusters_per_class=1)

df = pd.DataFrame(np.concatenate((X, y[:, np.newaxis]), axis=1),
                  columns=['x1', 'x2', 'y'])
print(df.head())
sns.pairplot(df, hue='y')
plt.show()
```
